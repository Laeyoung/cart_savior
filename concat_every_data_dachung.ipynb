```python
import urllib.request 
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
item_category_code = [100,200,300,400,500,600]
#(100:식량작물, 200:채소류, 300:특용작물, 400:과일류, 500:축산물, 600:수산물, default:100)
```

```python
df_list = []
for i in item_category_code:
    url = "http://www.kamis.or.kr/service/price/xml.do?action=dailyPriceByCategoryList" +\
    "&p_cert_key=bceaf385-9d34-4a75-9c6f-0607eb325485&p_cert_id=pje1740&p_returntype=json" +\
    "&p_product_cls_code=01" +\
    "&p_regday=2020-04-20" +\
    "&p_convert_kg_yn=N" +\
    "&p_item_category_code=" + str(i)
    response = urllib.request.urlopen(url)
    json_str = response.read().decode("utf-8")
    obj = json.loads(json_str)
    df = pd.DataFrame(obj['data']['item'])
    df_list.append(df.iloc[:, :9])
```
